**5. Identify and handle missing values appropriately.**

Check Missing Values

```python
df.isnull().sum()
```

Remove rows with missing values

```python
df_clean = df.dropna()
```

Fill missing values

For numeric:

```python
df['subscribers'] = df['subscribers'].fillna(df['subscribers'].median())
```

For categorical:

```python
df['category'] = df['category'].fillna('Unknown')
```

After handling:

```python
df.isnull().sum()
```
